```python
import pandas as pd
```

```python
df=pd.read_parquet("./Dataset/Online_retail_dec2010-dec2011.parquet",dtype_backend="pyarrow")
df.info()
```

## Data Cleaning

```python
df[df.isnull().any(axis=1)].head()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
c=df["Invoice"].str.startswith('C')
c.unique()
```

```python
df[c==True].head()
```

if Invoice startswith "C" it means that transction is a cancelled order

```python
cleandf=df[c!=True]
cleandf.head()
```

```python
del df
```

```python
cleandf[cleandf["Price"]==0].head()
```

price = 0 order is invalid (no use)

```python
cleandf=cleandf[cleandf["Price"]>0]
cleandf.head()
```

```python
cleandf["Revenue"]=cleandf["Quantity"]*cleandf["Price"]
cleandf.head()
```

```python
cleandf.convert_dtypes(dtype_backend="pyarrow")
cleandf.info()
```

```python
#cleandf.to_parquet("./Dataset/Cleaned_Online_retail_dec2010-dec2011.parquet")
```

```python
b1=cleandf.iloc[0:500,:]
```

```python
BasketData=b1.groupby('Invoice')["Description"].apply( lambda x: ' , '.join(x)).reset_index().drop('Invoice',axis=1).reset_index()

BasketData.head()
```

```python
BasketData.to_csv("productsdata500T.csv",index=False)
```
